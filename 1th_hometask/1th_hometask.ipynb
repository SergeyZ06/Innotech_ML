### Function "Reduce Memory Usage"

```python
def reduce_memory_usage(df):
    
    # СДЕЛАТЬ ПРОВЕРКУ type(df) == pd.DataFrame
    
    initial_memory_usage = df.memory_usage().sum() / 1024 / 1024
    print(f'Initial memory usage of dataframe:\t{initial_memory_usage:.3} Mb')
    
    for col in df.columns:
        col_type = df[col].dtype
        
        if col_type != 'object':
            c_min = df[col].min()
            c_max = df[col].max()
            
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            
            # Можно было бы сделать оптимизацию вплоть до типа "float16", однако данный тип данных,
            # как сообщается от сообщества аналитиков данных, плохо поддерживается некоторыми библиотеками
            
            if str(col_type)[:5] == 'float':
                if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                elif c_min > np.finfo(np.float64).min and c_max < np.finfo(np.float64).max:
                    df[col] = df[col].astype(np.float64)
        
        else:
            df[col] = df[col].astype('category')
    
    final_memory_usage = df.memory_usage().sum() / 1024 / 1024
    print(f'Final memory usage of dataframe:\t{final_memory_usage:.3} Mb')
    
    comparison = np.round(100 * (initial_memory_usage - final_memory_usage) / initial_memory_usage, 3)
    print(f'Memory usage has been decreased by:\t{comparison} %')
    
    return df
```

##### Данные признаки содержат пропуски, а также они наиболее важны для прогнозирования целевой переменной Medal.

##### Признаки Height и Weight можно было бы трансформировать в Индекс Массы Тела, но в таком случае возможна потеря информации.

##### Как выяснилось ранее, признак ID является уникальным идентификатором спортсмена, что может быть использовано для заполнения пропусков. Информация для заполнения пропусков может быть получена из других записей с идентичным ID.

```python
df_nulls = df_data[['ID', 'Age', 'Height', 'Weight']]
df_id_counts = pd.DataFrame(df_nulls['ID'].value_counts()).reset_index().rename({'ID': 'ID_counts'}, axis=1).rename({'index': 'ID'}, axis=1)
df_nulls = df_nulls.merge(df_id_counts, how='left', on='ID')

df_age_counts = pd.DataFrame(df_nulls['ID'][df_nulls['Age'].isna()].value_counts()).reset_index().rename({'ID': 'Age_null_counts'}, axis=1).rename({'index': 'ID'}, axis=1)
df_nulls = df_nulls.merge(df_age_counts, how='left', on='ID')

df_height_counts = pd.DataFrame(df_nulls['ID'][df_nulls['Height'].isna()].value_counts()).reset_index().rename({'ID': 'Height_null_counts'}, axis=1).rename({'index': 'ID'}, axis=1)
df_nulls = df_nulls.merge(df_height_counts, how='left', on='ID')

df_weight_counts = pd.DataFrame(df_nulls['ID'][df_nulls['Weight'].isna()].value_counts()).reset_index().rename({'ID': 'Weight_null_counts'}, axis=1).rename({'index': 'ID'}, axis=1)
df_nulls = df_nulls.merge(df_weight_counts, how='left', on='ID')

df_nulls = df_nulls.drop_duplicates()
df_nulls[(df_nulls['ID_counts'] > df_nulls['Age_null_counts']) | (df_nulls['ID_counts'] > df_nulls['Height_null_counts']) | (df_nulls['ID_counts'] > df_nulls['Weight_null_counts'])].head()
```

### Age, Height, Weight вывод:

##### Биологические характеристики стремятся к нормальному распределению, однако в случае недостаточно большой выборки из генеральной совокупности, может наблюдаться дисбаланс. Чтобы избежать влияния выбросов и дисбаланса используем медиану для заполнения пропусков.

##### Также можно было бы применить обработку с разбиением по регионам. Например, для спортсменов из Канады использовать медиану только канадских спортсменов.

### Team и NOC

##### Данные признаки содержат практически идентичную информацию, так как в Team записана страна, а в NOC - National Occupational Classification, что отражает тоже самое.

##### Данные признаки также копируют информацию друг друга. Ввиду того, что удобнее работать с Year и Season отдельно, от признака Games можно избавиться. Признак Season может быть переведён в категориальный.

### City, Event

##### Теоретически, данные признаки могут оказывать влияние на целевую переменную, однако в случае честных соревнований и объективного непредвзятого судейства все условия должны быть равны для спортсменов, независимо от места проведения события и самого события. Таким образом, данные признаки могут быть удалены в данный момент, но иметь потенциал для дальнейшего усовершенствования модели в дальнейшем.

### Sport

##### Очевидно, что данный признак должен оказывать влияние на целевую переменную. Например в случае, когда один спортсмен участвует в разных видах спорта и имеет разные преимущества и недостатки в разных видах спорта. Данный признак может быть переведён в категориальный.

### Medal

##### Целевая переменная. Должна быть переведена в категориальный тип, так как нельзя однозначно сказать, какое числовое соотношение будут иметь золотые, серебрянные и бронзовые медали.

## Data Preprocessing

##### Исчклюсения признаков, заполнение пропусков, обработка аномалий (заполнение пертентилями).

```python
df_data_preprocessed = df_data.copy()

df_data_preprocessed = df_data_preprocessed.drop(columns=['ID', 'Name', 'Team', 'Games', 'City', 'Event'])

df_data_preprocessed['Sex_M'] = False
df_data_preprocessed['Sex_M'][df_data_preprocessed['Sex'] == 'M'] = True
df_data_preprocessed = df_data_preprocessed.drop(columns=['Sex'])

df_data_preprocessed['Medal_Gold'] = False
df_data_preprocessed['Medal_Silver'] = False
df_data_preprocessed['Medal_Bronze'] = False
df_data_preprocessed['Medal_Gold'][df_data_preprocessed['Medal'] == 'Gold'] = True
df_data_preprocessed['Medal_Silver'][df_data_preprocessed['Medal'] == 'Silver'] = True
df_data_preprocessed['Medal_Bronze'][df_data_preprocessed['Medal'] == 'Bronze'] = True
df_data_preprocessed = df_data_preprocessed.drop(columns=['Medal'])


df_data_preprocessed['Age'] = df_data_preprocessed['Age'].fillna(value=df_data_preprocessed['Age'].median())
df_data_preprocessed['Age'][df_data_preprocessed['Age'] > df_data_preprocessed['Age'].quantile(q=0.999)] = df_data_preprocessed['Age'].quantile(q=0.999)

df_data_preprocessed['Height'] = df_data_preprocessed['Height'].fillna(value=df_data_preprocessed['Height'].median())

df_data_preprocessed['Weight'] = df_data_preprocessed['Weight'].fillna(value=df_data_preprocessed['Weight'].median())
df_data_preprocessed['Weight'][df_data_preprocessed['Weight'] > df_data_preprocessed['Weight'].quantile(q=0.95)] = df_data_preprocessed['Weight'].quantile(q=0.95)

df_data_preprocessed.head()
```
